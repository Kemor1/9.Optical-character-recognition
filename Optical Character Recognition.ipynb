# Import necessary libraries

```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

# 1. Getting an image

## 1.1 Load image

```python
img = np.array(Image.open('text_img.jpeg'))
```

## 1.2 Display image

```python
plt.imshow(img)
plt.show()
```

# 2. Preprocess image

## 2.1 Image normalization 

```python
img = cv2.normalize(img, np.zeros((img.shape[0], img.shape[1])), 0, 255, cv2.NORM_MINMAX)
```

```python
plt.imshow(img)
plt.show()
```

## 2.2 Image thresholding

```python
img = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)[1]
```

```python
plt.imshow(img)
plt.show()
```

# 3. Convert image to text

```python
print(pytesseract.image_to_string(img))
```
